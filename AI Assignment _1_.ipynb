```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
                                                    
#Question (2.a)
######################################

#Loading training data into memory
xi=pd.read_csv('hw1xtr.dat.txt',header=None)
yi=pd.read_csv('hw1ytr.dat.txt',header=None)



#Plotting training data
plt.plot(xi,yi,'o')
plt.title("Training Data")
plt.xlabel("Training Input")
plt.ylabel("Training Output")
plt.show()


#Loading test data into memory
xii=pd.read_csv('hw1xte.dat.txt',header=None)
yii=pd.read_csv('hw1yte.dat.txt',header=None)

#Plotting test data
plt.plot(xii,yii,'ro')
plt.title("Test Data")
plt.xlabel("Test Input")
plt.ylabel("Test Output")
plt.show()


```

```python
#Question (2.b)
######################################

#Creating a column vector of 1
one_vector=np.ones((len(xi),1))


#Adding column vector of 1
X=np.hstack((xi,one_vector))

#To find linear Regression 
w=(np.linalg.inv(X.T@X))@X.T@yi


#Predicting yi
yi_pred=(w[0][0]*xi)+w[0][1]

#Plotting Linear regression line and training data on same graph

a,b=zip(*sorted(zip(xi.values,yi_pred.values)))
plt.title("Training Data and Linear Regression Line")
plt.xlabel("Training Input")
plt.ylabel("Training Output")
plt.plot(a,b)
plt.plot(xi,yi,'yo')
plt.legend(["Linear Regression Line","Training Data"])
plt.show()


#ERROR on Training Data

avrgerr=(((((yi_pred)-(yi))**2).sum(axis = 0, skipna = True))/len(xi))
print("Average Training Error",avrgerr.values)



```

```python
#Question (2.c)
#Plotting Linear Regression line and test data on graph

#Predicting yi value
yi_pred=(w[0][0]*xii)+w[0][1]

a,b=zip(*sorted(zip(xii.values,yi_pred.values)))
plt.title("Test Data and Linear Regression Line")
plt.xlabel("Test Input")
plt.ylabel("Test Output")
plt.plot(a,b)
plt.plot(xii,yii,'o')
plt.legend(["Linear Regression Line","Test Data"])
plt.show()

#ERROR on Test Data

avrgerr=(((((yi_pred)-(yii))**2).sum(axis = 0, skipna = True))/len(xii))
print("Average Test Error",avrgerr.values)

```

```python
#Question (2.d)

#Implementing 2nd order Polynomial Regresiion
sec_ord_val=xi**2
sec_ord_val_test=xii**2


X=np.hstack((sec_ord_val,X))

w=(np.linalg.inv(X.T@X))@X.T@yi


yi_pred=((w[0][0]*sec_ord_val)+(w[0][1]*xi)+(w[0][2]))

#Plotting Polynomial regression line(2nd order) and training data on same graph


plt.xlabel("Training Input")
plt.ylabel("Training Output")
a,b=zip(*sorted(zip(xi.values,yi_pred.values)))
plt.plot(a,b)
plt.plot(xi,yi,'o')
plt.legend(["2nd order Polynomial Regression","Training Data"])
plt.title("Training Data and 2nd Order Polynomial Regression")
plt.show()



#ERROR on Training Data

avrgerr=(((((yi_pred)-(yi))**2).sum(axis = 0, skipna = True))/len(xi))
print("Average Training Error",avrgerr.values)

```

```python
#Plotting Polynomial Regression and test data on graph

yi_pred=((w[0][0]*sec_ord_val_test)+(w[0][1]*xii)+(w[0][2]))

plt.title("Test Data and 2nd order Polynomial Regression")
plt.xlabel("Test Input")
plt.ylabel("Test Output")
a,b=zip(*sorted(zip(xii.values,yi_pred.values)))
plt.plot(a,b)
plt.plot(xii,yii,'o')
plt.legend(["2nd order Polynomial Regression Line","Test Data"])
plt.show()

#ERROR on Test Data


avrgerr=(((((yi_pred)-(yii))**2).sum(axis = 0, skipna = True))/len(xii))
print("Average Test Error",avrgerr.values)



```

Yes 2nd Order Polynomial Regression is a better fit than the Linear Regression as it reduces the error.

```python
#Question (2.e)

#Implement 3rd order Polynomial Regresiion
third_ord_val=xi**3
third_ord_val_test=xii**3

X=np.hstack((third_ord_val,X))


w=(np.linalg.inv(X.T@X))@X.T@yi

yi_pred=((w[0][0]*third_ord_val)+(w[0][1]*sec_ord_val)+(w[0][2]*xi)+w[0][3])

#Plotting Polynomial regression line(3rd order) and training data on same graph

plt.xlabel("Training Input")
plt.ylabel("Training Output")
a,b=zip(*sorted(zip(xi.values,yi_pred.values)))
plt.plot(a,b)
plt.plot(xi,yi,'o')
plt.title("Training data and 3rd Order Polynomial Regression")
plt.legend(["3rd order Polynomial Regression Line","Training Data"])
plt.show()



#ERROR on Training Data



avrgerr=(((((yi_pred)-(yi))**2).sum(axis = 0, skipna = True))/len(xi))
print("Average Training Error",avrgerr.values)

```

```python
#Plotting Polynomial Regression and test data on graph

yi_pred=((w[0][0]*third_ord_val_test)+(w[0][1]*sec_ord_val_test)+(w[0][2]*xii)+w[0][3])

plt.xlabel("Test Input")
plt.ylabel("Test Output")
a,b=zip(*sorted(zip(xii.values,yi_pred.values)))
plt.plot(a,b)
plt.title("Test Data and 3rd Order Polynomial Regression")
plt.plot(xii,yii,'o')
plt.legend(["Polynomial Regression Line","Test Data"])
plt.show()

#ERROR on Test Data


avrgerr=(((((yi_pred)-(yii))**2).sum(axis = 0, skipna = True))/len(xii))
print("Average Test Error ",avrgerr.values)



```
